```r
options(repr.plot.width=15, repr.plot.height=15)
plot(hp, qsec)
abline(lm(qsec~hp),col='blue')
text(hp,qsec+0.3,rownames(mtcars),cex=1)
```

a)

```r
h <- hatvalues(slr)
print(h)
hbar <- mean(h)
high_lev <- which(h > 2 * hbar)
which(h > 2 * hbar)
mtcars[high_lev,]
options(repr.plot.width=15, repr.plot.height=15)
plot(hp, qsec)
abline(lm(qsec~hp),col='blue')
text(hp,qsec+0.3,rownames(mtcars),cex=1)
points(hp[high_lev],qsec[high_lev],col="orange",cex=2)
```

Two points whose leverage exceeds 2 x mean leverage h hat are Maserati Bora (observation #31) with h value of 0.27459288  and Ford Pantera L (observation #29) with h value of 0.12568847 

b)

```r
E_star <- rstudent(slr)
print(E_star)
y_outlier <- which(abs(E_star) > 2)
which(abs(E_star) > 2)
mtcars[y_outlier,]
options(repr.plot.width=15, repr.plot.height=15)
plot(hp, qsec)
abline(lm(qsec~hp),col='blue')
text(hp,qsec+0.3,rownames(mtcars),cex=1)
points(hp[y_outlier],qsec[y_outlier],col="blue",pch="X")
```

Only 1 point has studentized residuals that exceeds 2 in absolute value which is the Merc 230 (observation #9) with studentized residual of 4.02132955

c)

```r
n = length(qsec)
k = 1
D <- cooks.distance(slr)
D
high_cook <- which(D > 4/(n-k-1))
which(D > 4/(n-k-1))
print(4/(n-k-1))
mtcars[high_cook,]
options(repr.plot.width=15, repr.plot.height=15)
plot(hp, qsec)
abline(lm(qsec~hp),col='blue')
text(hp,qsec+0.3,rownames(mtcars),cex=1)
points(hp[high_cook],qsec[high_cook],col="red",pch="C",cex=3)
```

Only 1 point has a cook's distance value higher than the cutoff which is Merc 230 (observation #9) with cook's distance of 0.2801482

d)

A point is considered as influential if and only if its Cook's distance value is > the cutoff of 4/(n-k-1). Merc 230 is the only point whose Cook's distance of 0.2801482 is greater than the cutoff of 0.1333333
